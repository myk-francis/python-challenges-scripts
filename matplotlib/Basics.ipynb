```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Functional
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```
