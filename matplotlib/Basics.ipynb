```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
# Functional
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
#OBJECT ORIENTED
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGER PLOT')
axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
ax1.plot(x, y, color='red')
ax2.plot(x,y, color='red')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x, z, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x, y, color='blue')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (12, 2))

axes[0].plot(x, y, color='blue', lw=3, ls='--') 
axes[1].plot(x, z, color='red', lw=3, ls='-')
```
